```python
pop_data2 = pd.melt(frame=temp_pop,
                  id_vars=pop_indexes,
                  value_vars=pop_columns,
                  var_name="iYear",
                  value_name="iPopulation")
pop_data2['iYear'] = pop_data2['iYear'].apply(lambda yr: yr[-4:])
pop_data2.head()
```
